# Popularity Based Recommender

```python
book_with_rating = ratings_df.merge(books_df, on ="ISBN")
```

```python
book_with_rating
```

```python
num_rating_df = book_with_rating.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating': 'Num-Rating'}, inplace = True)
```

```python
num_rating_df
```

```python
avg_rating_df = book_with_rating.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'Avg-Rating'}, inplace = True)
```

```python
avg_rating_df
```

```python
popu_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
```

```python
popu_df
```

```python
popu_df = popu_df[popu_df['Num-Rating']>=250].sort_values('Avg-Rating', ascending = 0).head(50)
```

```python
popu_df.shape
```

```python
top_df = popu_df.merge(books_df, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Year-Of-Publication', 'Image-URL-L', 'Num-Rating', 'Avg-Rating']]
```

```python
top_df.reset_index(inplace = True, drop= True)
```

# Collaborative Based Recommendation

```python
book_with_rating
```

```python
# Finding Users that rated more than 200 books
x = book_with_rating.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
users = x[x].index
```

```python
users
```

```python
user_sort = book_with_rating[book_with_rating['User-ID'].isin(users)]
```

```python
user_sort
```

```python
# Finding books that have more than 50 ratings
y = user_sort.groupby('Book-Title').count()['Book-Rating'] >= 50
fam_books = y[y].index
```

```python
user_fam_df = user_sort[user_sort['Book-Title'].isin(fam_books)]
```

```python
user_fam_df
```

```python
pt = user_fam_df.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt.fillna(0, inplace = True)
```

```python
pt
```

```python
#pkl.dump(pt, open('pt_new.pkl', 'wb'))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_score = cosine_similarity(pt)
```

```python
sim_score.shape
```

```python
tri = np.where(pt.index == '1984')[0][0]
tri_lst = list(enumerate(sim_score[tri]))
arr = sorted(tri_lst, key=lambda x:x[1], reverse = True)[1:6]
for i in arr:
    print(i[0])
```

```python
def recommend(book_name):
    book_ind = np.where(pt.index == book_name)[0][0]
    book_sim_index = enumerate(sim_score[book_ind])
    book_sim_sorted = sorted(book_sim_index, key=lambda x:x[1], reverse = True)[1:6]
    for i in book_sim_sorted:
        book_info = user_fam_df[user_fam_df['Book-Title'] == pt.index[i[0]]].iloc[:1]
        book_info_list = []
        book_info_list.extend([book_info['Book-Title'].values[0], book_info['Book-Author'].values[0], book_info['Year-Of-Publication'].values[0], book_info['Image-URL-L'].values[0]])
        print(book_info_list)
    
    
```

```python
recommend('1984')
```
